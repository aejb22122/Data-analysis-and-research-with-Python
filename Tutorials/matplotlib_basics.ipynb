## Multiple line plots and basic customisations

```python
# Setting the variables
year = df['Year'].values
gdp_current = df['GDP_current_US'].values
fce_current = df['Final_consumption_expenditure_current_US'].values
fce_constant = df['Final_consumption_expenditure_constant 2010_US'].values

# Plotting
plt.plot(year, gdp_current, label = 'GDP (current USD)') # you can add customizations like colors, style of the line, and thickness of the lines as : color = 'k', linestyle ='--', linewidth = 3
plt.plot(year, fce_current, label = 'Final Consumption Expenditure (current USD)')
plt.plot(year, fce_constant, label = 'Final Consumption Expenditure (constant USD)')

# Annotating the plots
plt.annotate('2008 financial crisis', xy=(2008,0.6*1e12), #The point (x,y) to annotate
             xytext=(1993, 0.8*1e12), # The position (x, y) to place the text at
             arrowprops=dict(facecolor='black', shrink=0.05)) # The arrow will be based according to the diagonal made from the xy() and the xytext values ==> The point and the position

# Adding a vertical line
plt.axvline(2015, color = 'black', linestyle = '--', linewidth = 2, label = '2015 redefining method')

# Title and legends
plt.title('LDC indicators')
plt.legend()
plt.xlabel('Year')
plt.tight_layout(); # tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area.
```

## Bar graphs mixed with line plots

```python
# Setting the variables
year = df['Year'].values
gdp_current = df['GDP_current_US'].values
fce_current = df['Final_consumption_expenditure_current_US'].values
fce_constant = df['Final_consumption_expenditure_constant 2010_US'].values

# Plotting
plt.bar(year, gdp_current, color = 'gray', label = 'GDP (current USD)')
plt.plot(year, fce_current, color ='black', label = 'Final Consumption Expenditure (current USD)')
plt.plot(year, fce_constant, color = 'red', label = 'Final Consumption Expenditure (constant USD)')

# Title and legends
plt.title('LDC indicators \n (Mixing multiple chart types)')
plt.legend()

plt.tight_layout(); # tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area.
```

## Bar graphs side-by-side
<https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py>

```python
# Let's work with a subset of the data so that the graphs can be cleaner
data = df.loc[(df['Year'] > 2005)]
```

```python
# Setting the x values
x_indexes = np.arange(len(data['Year']))
# We will the add and substract from these x_indexes to have the bars side by side in the graphs
width = 0.25
```

```python
# Setting the bar y variables
y1 = data['GDP_current_US'].values
y2 = data['Final_consumption_expenditure_current_US'].values
y3 = data['Final_consumption_expenditure_constant 2010_US'].values
```

```python
the_years = list(data['Year'])
```

```python
# It's the same code as previously, just that we replaced the year with the x_indexes variable
# Plotting
plt.bar(x_indexes - width, y1, width = width, color='grey', label = 'GDP (current USD)') # width = largeur de la barre; on enlege width = 0.25 sur chaque barre
plt.bar(x_indexes, y2, width = width, color ='black', alpha = 0.85, label = 'Final Consumption Expenditure (current USD)') # on n'enleve pas le width ici parce que c'est la barre du millieu
plt.bar(x_indexes + width, y3, width = width, color='darkred', alpha = 0.75, label = 'Final Consumption Expenditure (constant USD)') # on ajoute le width ici

# Title and legends
plt.title("Least Developped Countries indicators ", fontsize=15, fontweight='bold')
plt.suptitle("Multiple bar charts", fontsize=10)

plt.xticks(ticks = x_indexes, labels = the_years)
plt.legend()

plt.tight_layout(); # tight_layout automatically adjusts subplot parametters so that the subplot(s) fits in to the figure area.
```

## Plotting with categorical variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset (if you don't have it loaded already)
titanic = sns.load_dataset('titanic')

# First change format from numeric to categorical (if it's not already)
titanic["sex"] = titanic["sex"].astype('category')

# Get the counts for each sex
sex_counts = titanic["sex"].value_counts()

# Plotting with matplotlib
plt.bar(sex_counts.index, sex_counts.values)  # Create the bar plot
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('Sex of Titanic Passengers')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```

```python
# Calculate survival rates by sex and class
survival_rates = titanic.groupby(['sex', 'class'], observed=True)['survived'].mean().reset_index()

# Create the bar plot
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figure size if needed
for s in survival_rates['sex'].unique():
    data = survival_rates[survival_rates['sex'] == s]
    ax.bar(data['class'], data['survived'], label=s)

# Set labels and title
ax.set_xlabel("Passenger Class")
ax.set_ylabel("Survival Rate (Mean)")
ax.set_title("Survival Rate by Sex and Passenger Class")
ax.legend(title="Sex")  # Add a legend to indicate sex
plt.show()
```
